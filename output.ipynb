```python
import pandas as pd
```

```python
df1 = pd.read_csv("KCLT.csv",index_col = "date")
df1.drop(columns=["actual_min_temp","actual_max_temp","average_min_temp","average_max_temp","record_min_temp","record_max_temp",
                 "record_min_temp_year","record_max_temp_year","average_precipitation","record_precipitation"], errors="ignore", inplace=True)
```

```python
df1.head()
```

```python
df1.columns = ["CLT_temp","CLT_precipitation"]
```

```python
df1.head()
```

```python
dataframe = df1
for city in ['KCQT', 'KHOU', 'KIND', 'KJAX',
                'KMDW', 'KNYC', 'KPHL', 'KPHX', 'KSEA']:
    df = pd.read_csv("%s.csv" % city, index_col = "date")
    df.drop(columns=["actual_min_temp","actual_max_temp","average_min_temp","average_max_temp","record_min_temp","record_max_temp",
                 "record_min_temp_year","record_max_temp_year","average_precipitation","record_precipitation"], errors="ignore", inplace=True)
    df.columns = ["%s_temp" % city,"%s_precipitation" % city]
    dataframe = pd.concat([dataframe, df], axis=1)
```

```python
dataframe
```

```python
dataframe.to_csv("output.csv")
```
